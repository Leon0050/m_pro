```python
import torch
```

```python
# 求所有元素积
```

```python
a = torch.tensor([[1,2],[3,4]])
result = torch.prod(a)
print(result)
```

```python
result_dim = torch.prod(a, dim=0) #沿着0维（行）向量计算
print(result_dim)
```

```python
result_keepdim = torch.prod(a, dim= 0, keepdim=True) #保持原始维度
print(result_keepdim)
```
