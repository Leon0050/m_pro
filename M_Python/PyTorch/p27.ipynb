```python
import torch
```

```python
w = torch.tensor([1.], requires_grad = True)
x = torch.tensor([2.], requires_grad = True)
for i in range(3):
    a = torch.add(x,w)
    b = torch.add(w,1)
    y = torch.mul(a,b)
    y.backward() # 自动执行反向传播并计算梯度（导数）存储到.grad中
    print(w.grad)
    # w.grad.zero_()
```
